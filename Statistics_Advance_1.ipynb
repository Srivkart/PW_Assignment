#Ques3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution?

```python
def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
    x (int): The value of the random variable (0 or 1)
    p (float): The probability of success (0 <= p <= 1)

    Returns:
    float: Probability P(X = x)
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 for Bernoulli distribution.")
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1.")

    return p**x * (1 - p)**(1 - x)

# Example usage:
print("P(X=1):", bernoulli_pdf(1, 0.7))
print("P(X=0):", bernoulli_pdf(0, 0.7))
```

#Ques 9.Write a Python function to calculate Z-scores from a dataset and plot them?

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculates Z-scores from the given dataset and plots them.

    Parameters:
    data (list or numpy array): The input dataset.
    """
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data)

    # Calculate Z-scores
    z_scores = (data - mean) / std

    # Plot Z-scores
    plt.figure(figsize=(8, 5))
    plt.plot(z_scores, marker='o', linestyle='-', color='blue')
    plt.axhline(0, color='black', linestyle='--', label='Mean (Z=0)')
    plt.axhline(1, color='green', linestyle='--', label='1 Std Dev')
    plt.axhline(-1, color='green', linestyle='--')
    plt.title("Z-Scores of the Dataset")
    plt.xlabel("Data Point Index")
    plt.ylabel("Z-Score")
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.legend()
    plt.show()

    return z_scores

# Example usage
data = np.random.normal(loc=50, scale=10, size=100)
z_scores = calculate_and_plot_z_scores(data)
```

#Ques14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std_dev):
    """
    Calculate the Z-score for a given data point.

    Parameters:
    x : float - Data point
    mean : float - Mean of the dataset
    std_dev : float - Standard deviation of the dataset

    Returns:
    float - Z-score
    """
    return (x - mean) / std_dev

def plot_standard_normal_with_z(z):
    """
    Plot standard normal distribution and mark the Z-score.

    Parameters:
    z : float - Z-score to be marked on the curve
    """
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, loc=0, scale=1)

    plt.figure(figsize=(10, 5))
    plt.plot(x, y, label='Standard Normal Distribution', color='blue')
    plt.fill_between(x, 0, y, where=(x <= z), color='lightblue', alpha=0.6, label=f'Area â‰¤ Z={z:.2f}')
    plt.axvline(z, color='red', linestyle='--', linewidth=2, label=f'Z = {z:.2f}')
    plt.title("Z-score Compared to Standard Normal Distribution")
    plt.xlabel("Z")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()

# Example input
x_value = 85
mean = 70
std_dev = 10

# Calculate Z-score
z = calculate_z_score(x_value, mean, std_dev)
print(f"Z-score for x = {x_value} is: {z:.2f}")

# Visualize
plot_standard_normal_with_z(z)
```

#Ques15. Implement hypothesis testing using Z-statistics for a sample dataset?

```python
import numpy as np
from scipy.stats import norm

def z_test(sample_mean, population_mean, population_std, n, alpha=0.05):
    # Step 1: Calculate the Z-statistic
    z = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    # Step 2: Calculate critical z-value for two-tailed test
    z_critical = norm.ppf(1 - alpha / 2)

    # Step 3: Calculate p-value
    p_value = 2 * (1 - norm.cdf(abs(z)))

    # Step 4: Print results
    print(f"Z-statistic: {z:.4f}")
    print(f"Critical Z-value (Â±): {z_critical:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Step 5: Conclusion
    if abs(z) > z_critical:
        print("Result: Reject the null hypothesis (evidence of a significant difference).")
    else:
        print("Result: Fail to reject the null hypothesis (no significant difference).")

# Example parameters
sample_mean = 980
population_mean = 1000
population_std = 80
sample_size = 50
alpha = 0.05

# Run the test
z_test(sample_mean, population_mean, population_std, sample_size, alpha)
```

#Ques 16. Create a confidence interval for a dataset using Python and interpret the result?

```python
import numpy as np
from scipy.stats import norm

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for a dataset.

    Parameters:
    data (array-like): Sample data
    confidence (float): Confidence level (e.g., 0.95 for 95%)

    Returns:
    tuple: (lower_bound, upper_bound)
    """
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)  # sample std deviation
    z = norm.ppf(1 - (1 - confidence) / 2)

    margin_of_error = z * (std_dev / np.sqrt(n))
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    print(f"Sample Mean: {mean:.2f}")
    print(f"{int(confidence * 100)}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

    return lower_bound, upper_bound

# Example: generate synthetic dataset
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=30)

# Calculate and interpret 95% confidence interval
confidence_interval(data, confidence=0.95)
```

#Ques 21. Write a Python program to compare the standard deviations of two datasets and visualize the difference?

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate two datasets
np.random.seed(42)

# Dataset A: smaller standard deviation
data_a = np.random.normal(loc=50, scale=5, size=1000)

# Dataset B: larger standard deviation
data_b = np.random.normal(loc=50, scale=15, size=1000)

# Step 2: Calculate standard deviations
std_a = np.std(data_a, ddof=1)
std_b = np.std(data_b, ddof=1)

print(f"Standard Deviation of Dataset A: {std_a:.2f}")
print(f"Standard Deviation of Dataset B: {std_b:.2f}")

# Step 3: Plot the histograms
plt.figure(figsize=(10, 5))
plt.hist(data_a, bins=30, alpha=0.6, label=f'Dataset A (std = {std_a:.2f})', color='skyblue', edgecolor='black')
plt.hist(data_b, bins=30, alpha=0.6, label=f'Dataset B (std = {std_b:.2f})', color='salmon', edgecolor='black')
plt.title('Comparison of Standard Deviations')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.axvline(np.mean(data_a), color='blue', linestyle='--', linewidth=2, label='Mean A')
plt.axvline(np.mean(data_b), color='red', linestyle='--', linewidth=2, label='Mean B')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

#Ques22. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution?

```python
import numpy as np

# Step 1: Generate a dataset from a normal distribution
np.random.seed(0)
data = np.random.normal(loc=100, scale=15, size=1000)  # mean=100, std=15, n=1000

# Step 2: Calculate Range
data_range = np.max(data) - np.min(data)

# Step 3: Calculate IQR
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

# Output results
print(f"Range: {data_range:.2f}")
print(f"Q1 (25th percentile): {q1:.2f}")
print(f"Q3 (75th percentile): {q3:.2f}")
print(f"IQR (Interquartile Range): {iqr:.2f}")
```

#Ques 23. Implement Z-score normalization on a dataset and visualize its transformation?

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate a synthetic dataset
np.random.seed(1)
data = np.random.normal(loc=100, scale=20, size=1000)

# Step 2: Z-score normalization
mean = np.mean(data)
std = np.std(data)
z_scores = (data - mean) / std

# Step 3: Plot original and normalized data side-by-side
plt.figure(figsize=(12, 5))

# Original data histogram
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title("Original Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.axvline(mean, color='red', linestyle='--', label=f"Mean = {mean:.2f}")
plt.legend()

# Normalized data histogram
plt.subplot(1, 2, 2)
plt.hist(z_scores, bins=30, color='lightgreen', edgecolor='black')
plt.title("Z-score Normalized Data")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.axvline(0, color='red', linestyle='--', label="Mean = 0")
plt.legend()

plt.tight_layout()
plt.show()
```

#Ques24. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution?

```python
import numpy as np
from scipy.stats import skew, kurtosis

def compute_skewness_kurtosis(data):
    """
    Computes and prints skewness and kurtosis of the given dataset.

    Parameters:
    data (array-like): Numerical dataset
    """
    data = np.array(data)
    skew_val = skew(data)
    kurt_val = kurtosis(data, fisher=False)  # Set fisher=True for excess kurtosis

    print(f"Skewness: {skew_val:.4f}")
    print(f"Kurtosis: {kurt_val:.4f} (normal dist = 3.0)")

# Generate data from a normal distribution
np.random.seed(42)
normal_data = np.random.normal(loc=0, scale=1, size=1000)

# Compute skewness and kurtosis
compute_skewness_kurtosis(normal_data)
```
