#Ques5. What are iterators in Python and how do they differ from iterables?
- **Answer**

  **Iterable:**
 An iterable is any Python object that can be looped over (i.e., used in a for loop).

 It implements the __iter__() method and returns an iterator.

 - **Examples of iterables:** list, tuple, string, dict, set.

  **Iterator:**

   An iterator is an object that represents a stream of data, returning one element at a time using the next() function.
   It must implement both:

  __iter__() → returns itself

  __next__() → returns the next item (or raises StopIteration)



```python
#Example
# Iterable
nums = [10, 20, 30]  # list is iterable

# Convert to iterator
it = iter(nums)  # now 'it' is an iterator

print(next(it))  # Output: 10
print(next(it))  # Output: 20
print(next(it))  # Output: 30
# print(next(it))  # Raises StopIteration
```

#Ques6. Explain the concept of generators in Python and how they are defined.

- **Answer:**

 **Generator:**

   A generator is a special type of iterator that yields values one at a time, allowing you to generate items on the fly without storing the entire sequence in memory.

  **How are Generators Defined**

   Generators are typically created in two ways:

 - **Generator Functions:**

   Defined like a normal function but use the yield keyword to produce values one at a time.

 - **Generator Expressions:**
  Similar to list comprehensions but with parentheses () instead of square brackets [].


```python
#Example of Generator
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1

# Using the generator
for number in count_up_to(5):
    print(number)
```

#Ques 7. What are the advantages of using generators over regular functions?

- **Answer:**

 - 1. Memory Efficiency

     Generators produce items one at a time, so they don’t store the entire sequence in memory. Regular functions usually return a complete list, which can be large and consume a lot of memory.

 - 2. Lazy Evaluation

     Generators yield values only when requested, which is useful for processing large or infinite data streams. Regular functions compute and return all results at once.

 - 3. Improved Performance

     Because generators don’t build the whole list in memory, they can start producing results faster and handle large datasets without slowing down.

 - 4. Simpler Code for Iteration

     Generators can replace complex iterator classes with clean, simple functions using yield.

```python
#Example: Regular Function (returns list):
def get_squares(n):
    result = []
    for i in range(n):
        result.append(i * i)
    return result

squares = get_squares(5)
print(squares)  # Output: [0, 1, 4, 9, 16]

#Generator Function (yields values):
def generate_squares(n):
    for i in range(n):
        yield i * i

for square in generate_squares(5):
    print(square)
```

#Ques 8. What is a lambda function in Python and when is it typically used?

- **Answer:**

 - **Definition:**
A lambda function is a small, anonymous (unnamed) function defined with the keyword lambda.
It can have any number of arguments but only one expression, which is evaluated and returned automatically.

 - **When is it Typically Used**
For short, simple functions that are used temporarily.Commonly passed as arguments to functions like map(), filter(), and sorted().
When defining a full function with def feels too verbose for a small operation.

 - **Syntax**

   lambda arguments: expression


```python
#Example
# Regular function to add two numbers
def add(x, y):
    return x + y

print(add(3, 5))  # Output: 8

# Equivalent lambda function
add_lambda = lambda x, y: x + y
print(add_lambda(3, 5))  # Output: 8
```

#Ques 9. Explain the purpose and usage of the `map()` function in Python.
 - **Answer:**

    - **map()**

   map() applies a given function to each item of an iterable (like a list) and returns a map object (an iterator) with the results.
   It’s used to transform or process all items in a collection without writing explicit loops.

**Syntax:**

map(function, iterable, ...)


```python
#Example
# Function to square a number
def square(x):
    return x * x

numbers = [1, 2, 3, 4, 5]

# Using map to apply 'square' to each number
squared_numbers = map(square, numbers)

# Convert map object to list to see results
print(list(squared_numbers))  # Output: [1, 4, 9, 16, 25]
```

#Ques 10. What is the difference between map(), reduce(), and filter() functions in Python?
 - **Answer:**
  - **map()**

     The map() function applies a given function to every item in an iterable (like a list) and returns an iterator with the transformed items. It’s useful when you want to perform the same operation on each element of a collection.

  - **filter()**

     The filter() function applies a test function to each item of an iterable and returns an iterator with only those items for which the test function returns True. It’s used to select elements based on a condition.

  - **reduce()**

     The reduce() function (from the functools module) applies a function cumulatively to the items of an iterable, reducing it to a single value. The function takes two arguments and combines the iterable’s elements step-by-step.


```python
#Example of map
nums = [1, 2, 3, 4]
squared = map(lambda x: x ** 2, nums)
print(list(squared))  # Output: [1, 4, 9, 16]

#Example of Filter
nums = [1, 2, 3, 4, 5, 6]
evens = filter(lambda x: x % 2 == 0, nums)
print(list(evens))  # Output: [2, 4, 6]

#Example of reduce
from functools import reduce

nums = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, nums)
print(product)  # Output: 24
```

#Ques1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

```python
def sum_of_evens(numbers):
    """
    Takes a list of numbers and returns the sum of all even numbers.
    """
    return sum(num for num in numbers if num % 2 == 0)

# Main program
if __name__ == "__main__":
    # Get input from user
    input_str = input("Enter a list of numbers separated by spaces: ")

    # Convert input string to a list of integers
    number_list = list(map(int, input_str.split()))

    # Call the function
    even_sum = sum_of_evens(number_list)

    # Print the result
    print("Sum of even numbers:", even_sum)
```

#Ques 4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    """
    Checks if a number is prime.
    Returns True if prime, False otherwise.
    """
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Main program: check numbers from 1 to 200
print("Prime numbers from 1 to 200:")
for number in range(1, 201):
    if is_prime(number):
        print(number, end=' ')
```

#Ques 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python
class FibonacciIterator:
    """
    Iterator class to generate Fibonacci sequence up to a specified number of terms.
    """
    def __init__(self, num_terms):
        self.num_terms = num_terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.num_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b

# Main program
if __name__ == "__main__":
    # Get number of terms from the user
    n = int(input("Enter the number of Fibonacci terms: "))

    print("Fibonacci sequence:")
    for num in FibonacciIterator(n):
        print(num, end=' ')
```

#Ques 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(max_exponent):
    """
    Generator function that yields powers of 2 from 0 to max_exponent.
    """
    for exponent in range(max_exponent + 1):
        yield 2 ** exponent

# Main program
if __name__ == "__main__":
    # Get input from user
    max_exp = int(input("Enter the maximum exponent: "))

    print(f"Powers of 2 from 0 to {max_exp}:")
    for value in powers_of_two(max_exp):
        print(value, end=' ')
```

# Ques 7.  Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_line_by_line(filename):
    """
    Generator function that reads a file line by line and yields each line.
    """
    try:
        with open(filename, 'r') as file:
            for line in file:
                yield line.rstrip('\n')  # Remove newline character
    except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found.")

# Main program
if __name__ == "__main__":
    # Get file name from user
    file_name = input("Enter the filename to read: ")

    print(f"\nContents of '{file_name}':")
    for line in read_file_line_by_line(file_name):
        print(line)
```

#ques 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
def celsius_to_fahrenheit(c):
    """
    Convert Celsius temperature to Fahrenheit.
    """
    return (c * 9/5) + 32

# Sample list of temperatures in Celsius
celsius_temps = [0, 20, 37, 100]

# Use map() to convert all temperatures to Fahrenheit
fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))

print("Celsius temperatures:", celsius_temps)
print("Fahrenheit temperatures:", fahrenheit_temps)
```

#ques 10.  Create a Python program that uses filter() to remove all the vowels from a given string.

```python
def is_not_vowel(char):
    """
    Returns True if the character is not a vowel, False otherwise.
    """
    return char.lower() not in 'aeiou'

# Get input from the user
input_str = input("Enter a string: ")

# Use filter() to remove vowels
filtered_chars = filter(is_not_vowel, input_str)

# Join the filtered characters back into a string
result = ''.join(filtered_chars)

print("String without vowels:", result)
```
