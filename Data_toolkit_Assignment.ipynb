#Ques4.  How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame (replace with your data)
df = pd.DataFrame({
    'Age': [25, 30, 35, 40, 45, 50, 55, 60],
    'Income': [40000, 50000, 60000, 70000, 80000, 90000, 100000, 110000],
    'Spending': [20000, 25000, 30000, 35000, 40000, 45000, 50000, 55000],
    'Savings': [5000, 7000, 9000, 11000, 13000, 15000, 17000, 19000]
})

corr_matrix = df.corr()
print(corr_matrix)

plt.figure(figsize=(10, 8))

# Create the heatmap with annotations
heatmap = sns.heatmap(
    corr_matrix,
    annot=True,  # Show correlation values
    fmt=".2f",  # Format to 2 decimal places
    cmap='coolwarm',  # Color scheme
    center=0,  # Center the colormap at 0
    linewidths=0.5,  # Add lines between cells
    linecolor='white',  # Line color
    square=True  # Make cells square
)

# Add title
plt.title('Correlation Matrix Heatmap', fontsize=16, pad=20)

# Rotate x-axis labels for better readability
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=45, horizontalalignment='right')

# Show the plot
plt.tight_layout()
plt.show()

# Create a mask to show only correlations above 0.5
mask = np.abs(corr_matrix) < 0.5
sns.heatmap(corr_matrix, mask=mask, annot=True, cmap='viridis')

# Use clustermap to group similar correlations
sns.clustermap(corr_matrix, annot=True, cmap='coolwarm', figsize=(10, 8))

# Add asterisks for significant correlations
annot = corr_matrix.copy().astype(str) # Start with string representation of the matrix

# Create masks using the original numeric corr_matrix
mask_three_stars = corr_matrix > 0.7
mask_two_stars = (corr_matrix > 0.5) & (corr_matrix <= 0.7)

# Apply asterisks based on masks
annot[mask_three_stars] = annot[mask_three_stars] + "***"
annot[mask_two_stars] = annot[mask_two_stars] + "**"

# Create the heatmap with the modified annot DataFrame
plt.figure(figsize=(10, 8)) # Add a new figure for this plot
sns.heatmap(corr_matrix, annot=annot, fmt='', cmap='coolwarm', center=0, linewidths=0.5, linecolor='white', square=True) # Use the original corr_matrix for colors and fmt, and annot for text
plt.title('Correlation Matrix Heatmap with Significance Markers', fontsize=16, pad=20)
plt.xticks(rotation=45, horizontalalignment='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

#Ques6. Create a DataFrame and add a new column based on an existing column?

```python
import pandas as pd
import numpy as np

# Create the initial DataFrame with Name, Age, Salary
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'Salary': [50000, 60000, 70000, 80000]
}
df = pd.DataFrame(data)

# Add a new column based on existing columns
df['Bonus'] = df['Salary'] * 0.1  # 10% bonus

print("DataFrame with new Bonus column:")
print(df)

# Add a column with conditional logic
df['Age_Group'] = df['Age'].apply(lambda x: 'Young' if x < 30 else 'Senior' if x >= 40 else 'Mid')

print("\nDataFrame with Age Group:")
print(df)

# Add a column combining multiple columns
df['Total_Compensation'] = df['Salary'] + df['Bonus']

print("\nDataFrame with Total Compensation:")
print(df)

# Create and modify DataFrame in one chain (This overwrites df)
df = (
    pd.DataFrame({
        'Product': ['A', 'B', 'C', 'D'],
        'Price': [10, 20, 15, 25],
        'Quantity': [100, 50, 75, 60]
    })
    .assign(Revenue=lambda x: x['Price'] * x['Quantity'])
    .assign(Profit=lambda x: x['Revenue'] * 0.2)
)

print("\nProduct DataFrame with calculated columns:")
print(df)

# Recreate the DataFrame with 'Name' column here
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'Salary': [50000, 60000, 70000, 80000]
}
df = pd.DataFrame(data)
# Re-add the columns that were present before the overwrite if needed for subsequent operations
df['Bonus'] = df['Salary'] * 0.1
df['Age_Group'] = df['Age'].apply(lambda x: 'Young' if x < 30 else 'Senior' if x >= 40 else 'Mid')
df['Total_Compensation'] = df['Salary'] + df['Bonus']


# Now add column with string manipulation
df['Initial'] = df['Name'].str[0]  # Get first letter of each name

print("\nDataFrame with Initials:")
print(df)

# Add column with numpy.where for conditional logic
df['High_Earner'] = np.where(df['Salary'] > 65000, 'Yes', 'No')

print("\nDataFrame with High Earner flag:")
print(df)

# Add multiple columns in one operation
df[['Salary_Monthly', 'Bonus_Monthly']] = df[['Salary', 'Bonus']].div(12).round(2)

print("\nDataFrame with monthly calculations:")
print(df)
```

#Ques7. Write a program to perform element-wise multiplication of two NumPy arrays?

```python
import numpy as np

# Method 1: Basic element-wise multiplication
def elementwise_multiply(arr1, arr2):
    """
    Performs element-wise multiplication of two numpy arrays

    Parameters:
    arr1 (numpy.ndarray): First input array
    arr2 (numpy.ndarray): Second input array

    Returns:
    numpy.ndarray: Result of element-wise multiplication
    """
    return arr1 * arr2

# Method 2: Using numpy.multiply() function
def numpy_multiply(arr1, arr2):
    """
    Performs element-wise multiplication using np.multiply()

    Parameters:
    arr1 (numpy.ndarray): First input array
    arr2 (numpy.ndarray): Second input array

    Returns:
    numpy.ndarray: Result of element-wise multiplication
    """
    return np.multiply(arr1, arr2)

# Create sample arrays
array1 = np.array([[1, 2, 3], [4, 5, 6]])
array2 = np.array([[7, 8, 9], [10, 11, 12]])

# Perform element-wise multiplication
result1 = elementwise_multiply(array1, array2)
result2 = numpy_multiply(array1, array2)

# Verify both methods produce same result
assert np.array_equal(result1, result2)

# Print results
print("Array 1:")
print(array1)
print("\nArray 2:")
print(array2)
print("\nElement-wise multiplication result (* operator):")
print(result1)
print("\nElement-wise multiplication result (np.multiply):")
print(result2)
```

#Ques8. Create a line plot with multiple lines using Matplotlib?

```python
import matplotlib.pyplot as plt
import numpy as np

# Create sample data
x = np.linspace(0, 10, 100)  # X-axis values (0 to 10 in 100 steps)
y1 = np.sin(x)                # First line: sine wave
y2 = np.cos(x)                # Second line: cosine wave
y3 = np.sin(x + 0.5)          # Third line: phase-shifted sine wave

# Create figure and axis
plt.figure(figsize=(10, 6))   # Set figure size (width, height)

# Plot multiple lines with different styles
plt.plot(x, y1, label='Sine Wave', color='blue', linestyle='-', linewidth=2)
plt.plot(x, y2, label='Cosine Wave', color='red', linestyle='--', linewidth=2)
plt.plot(x, y3, label='Phase-Shifted Sine', color='green', linestyle=':', linewidth=2)

# Add title and labels
plt.title('Multiple Line Plot Example', fontsize=14, pad=20)
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)

# Add legend
plt.legend(fontsize=10, framealpha=1, shadow=True)

# Customize grid and ticks
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```

#Ques10. Create a histogram using Seaborn to visualize a distribution

#Ques11. Perform matrix multiplication using NumPy

```python
import numpy as np

# Define two matrices
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

# Perform matrix multiplication
result = np.matmul(A, B)

# Alternatively, you can also use the @ operator (Python 3.5+)
# result = A @ B

# Display the result
print("Matrix A:")
print(A)
print("\nMatrix B:")
print(B)
print("\nMatrix Multiplication (A * B):")
print(result)
```
